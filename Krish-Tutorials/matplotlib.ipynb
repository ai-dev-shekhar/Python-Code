```python
import matplotlib.pyplot as plt
%matplotlib inline
# no need to write plt.show(), the notebook is rendered so as to display the charts inline
```

```python
import numpy as np
import pandas as pd
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
# Scatter Plot
plt.scatter(x,y,c='b') # c='g' - color of points is green
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('----a 2-D graph----')
plt.savefig('2dGraph.png') # save the plot in local
```

```python
y = x*x
```

```python
# Line plot
plt.plot(x,y,'y*', linestyle = 'solid', linewidth=2, markersize=12)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line graph')
```

```python
# Creating subplots
plt.subplot(2,2,1) # the plot is to divided into 2x2 matrix i.e consists of 2 rows and 2 columns; the final parameter is the position of the next plot
plt.plot(x,y,'g*') # the points are green color and * shaped  
plt.subplot(2,2,2)
plt.plot(x,y,'k--')# the points are black color and -- shaped
plt.subplot(2,2,3)
plt.plot(x,y,'yo') # the points are yellow colored and o shaped
plt.subplot(2,2,4)
plt.plot(x,y,'c^') # the points are cyan colored and ^ shaped
plt.savefig('subplots.png')
```

```python
# Bar plot
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y,color='c')
plt.bar(x2, y2, color = 'b')
plt.title('A simple Bar-Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
# Histograms
arr2 = np.random.randint(1,100,15)
print(arr2)
plt.hist(arr2, color='y', bins=20)
plt.title('A simple Histogram')
```

```python
# Boxplot
box_data = [np.random.normal(0, std, 100) for std in range(1,4)]
# print(box_data)

# Rectangular box-plot
plt.boxplot(box_data, vert=True, patch_artist=True) # vert=True - creates vertical boxplots; patch_artist=True - boxplots are filled
plt.show()
```

```python
# Pie-chart
labels = ['Python', 'C++', 'Golang', 'Java', 'Ruby']
sizes = [215,135,180,206,76]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'grey']
explode = (0.5,0,0.1,0,0) # Explode 1st and 3rd slice, and by how much is indicated by the number after decimal

#plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True)
plt.axis('equal')
```
