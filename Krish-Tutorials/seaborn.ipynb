```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.info()
# df.describe()
```

## Numerical Plots

### Correlation with Heatmap
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrx b/w 2 discrete dimensions or event types.

```python
df_num = df[df.select_dtypes(include=['float64','int64']).columns] # selecting numerical columns
df_num.corr()
# df.select_dtypes(include=['float64','int64']).corr()
```

```python
sns.heatmap(df_num.corr())
```

### JointPlot
A join/joint plot allows to study the relationship b/w 2 numeric variables. The central chart displays the connection. It is usually a scatter plot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x = 'tip', y = 'total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data=df, kind='reg')
# kind = 'reg' creates a probability density function with the histogram. Also creates a regression/best-fit line for the scatter plot
```

### Pair Plot
A pair/scatter plot is the one in which multiple scatter plots are arranged in a grid to visualize the relationships between pairs of variables in a dataset. Pair plots are particularly useful for exploring the relationships between multiple variables simultaneously

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex') # separates based on the categories mentioned in the column provided for hue
```

### Dist Plot
Dist plot helps us to check the distribution of the columns feature

```python
sns.displot(df['tip'], kde=True)
```

## Categorical Plots

### Count Plot
We need to provide either x axis or y-axis because one of the axes will always be count

```python
sns.countplot(data=df,x='sex',hue='smoker',stat='count', orient='h')
```

```python
sns.countplot(data=df, x='day', orient='h')
```

```python
# Bar plot
sns.barplot(x='total_bill', y='sex', data=df)
```

### Box plot
A box and whisker plot(also called as boxplot) is a graph that presents information from a 5-number summary

```python
sns.boxplot(x='day', y='total_bill', hue='sex', data=df, palette='rainbow')
```

### Violin Plot
Violin plot helps us to see both the distribution in terms of kernel density estimation and the box plot

```python
sns.violinplot(data=df, x='total_bill', y='day')#, hue='smoker', palette='rainbow')
```
